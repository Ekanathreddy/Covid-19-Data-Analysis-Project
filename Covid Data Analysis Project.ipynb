```python
# Dropping the unwanted columns from the covid_19_india dataset
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%d/%m/%y')
```

```python
covid_df.head()
```

```python
# Active cases
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured']+covid_df['Deaths'])
```

```python
covid_df.head()
```

```python
# Creating pivot table summing all the confirmed, deaths, cured cases by State/UnionTerritory
statewise = pd.pivot_table(covid_df, values = ['Confirmed', 'Deaths', 'Cured'], index = 'State/UnionTerritory', aggfunc = max)
```

```python
statewise
```

```python
statewise['Recovery Rate'] = statewise['Cured']*100 / statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100 / statewise['Confirmed']
```

```python
# sorting
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

```python
statewise
```

```python
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
# Top 10 states based on the active cases
top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by = ['Active_cases'], ascending = False).reset_index()
# .max() returns maximum value in a collection and [] for selecting specific columns
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most active cases in India", size = 20)
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'red')
# here x is specified as in the dataframe (do not give you wished name here), similary y to
# iloc[:10] selects rows from index 0 to 10
# here states are overlapped so, look into the next cell
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active cases in India", size = 18)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = 'Active_cases', x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'red')
plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
# top states with highetst deaths
top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending=False).reset_index()
fig = plt.figure(figsize = (18,5))
plt.title("Top 10 states with most deaths", size=18)
ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor='blue')
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
# Growth Trend
fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka','Kerala','Tamil Nadu','Utter Pradesh'])], x = 'Date', y = 'Active_cases', hue = 'State/UnionTerritory')
plt.title("Top 5 affected states in india", size = 16)
```

```python
# Remove the state where state=india
vaccine = vaccine_df[vaccine_df['State'] != "India"] # or vaccine = vaccine_df[vaccine_df.State != "India"]
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'}, inplace = True)
```

```python
vaccine.head()
```

```python
# Most vaccinated state
max_vac = vaccine.groupby(by="State")['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values(by=['Total'],ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated States in India", size=20)
x = sns.barplot(data = max_vac.iloc[:10], y=max_vac.Total, x = max_vac.index, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
# Find the states with least individuals vaccinated and convert that table into a barplot
min_vac = vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values(by=['Total'],ascending=True)[:5]
min_vac
```

```python
fig = plt.figure(figsize=(15,5))
plt.title("Top 5 States with least individuals vaccinated", size=20)
ax = sns.barplot(data=min_vac.iloc[:5], y = min_vac.Total, x = min_vac.index, linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
